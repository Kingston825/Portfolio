```python
# create a df that we will use for our x and y values
finaldf = df.drop(columns=['TWF','HDF','PWF','OSF','RNF','Product ID','UDI'])
```

```python
x = finaldf.drop(columns=['Machine failure'])
y = finaldf['Machine failure']
```

```python
x = pd.get_dummies(x)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
```

```python
x_train,y_train = smote.fit_resample(x_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.transform(x_test)
```

```python
#train classification model using stacking(log reg, knn, decisiton tree, svm, 
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import StackingClassifier
```

```python
logr = LogisticRegression()
dt = DecisionTreeClassifier()
knn = KNeighborsClassifier()


```

```python
stack = StackingClassifier(estimators=[('logr',logr),('dt',dt),('knn',knn)])
```

```python
stack.fit(x_train_scaled,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = stack.predict(x_test_scaled)
```

```python
acc = accuracy_score(y_test, y_pred)
```

```python
#.963
print(acc)
```

```python

#recall
```

```python
# import packages 
from sklearn.metrics import roc_auc_score

# calculate AUC
roc_auc_score(y_test,y_pred)

```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import f1_score

f1_score(y_test,y_pred,average='binary')

```

```python
from sklearn.metrics import precision_score

precision = precision_score(y_test, y_pred)

print(f"Precision Score: {precision}")
```

```python
from sklearn.metrics import recall_score

recall = recall_score(y_test, y_pred)

print(f"recall Score: {recall}")
```

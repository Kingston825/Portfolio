```python
import pandas as pd
import numpy as np
df = pd.read_csv('labeledTrainData.tsv',sep='\t')
df.head(20)
```

```python
import pandas as pd
import re
import string
import nltk

stopwords = nltk.corpus.stopwords.words('english')
ps = nltk.PorterStemmer()
```

```python
def clean_text(txt):
    txt = "".join([c for c in txt if c not in string.punctuation])
    tokens = re.split('\W+', txt)
    txt = [ps.stem(word) for word in tokens if word not in stopwords]
    return txt
```

```python
df.head()
x = df['review']
y = df['sentiment']
```

## do train test split of data then make tfidf of each set only fitting one

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2)
```

```python
len(x_train) , len(x_test)
```

Fit and apply the tf-idf vectorization to the training set.


```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python

tfidf = TfidfVectorizer(analyzer=clean_text)
x_train = tfidf.fit_transform(x_train)
```

Apply but DO NOT FIT the tf-idf vectorization to the test set (Why?).


```python

x_test = tfidf.transform(x_test)
```

```python
x_train.shape[0] , x_test.shape[0]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression().fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy: ", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix

confusion = confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import precision_score
```

```python
precision_score(y_test, y_pred)
```

```python
from sklearn.metrics import recall_score
```

```python
 recall_score(y_test, y_pred)
```

```python
from sklearn.metrics import f1_score
```

```python
f1_score(y_test, y_pred)
```

```python
#### Create a Roc curve
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
y_pred_proba = model.predict_proba(x_test)[::,1]
```

```python
from matplotlib import pyplot as plt
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
```

```python
plt.plot(fpr,tpr)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

#### regenerate clean data set and create out tfidf of the data

```python
df = pd.read_csv('labeledTrainData.tsv',sep='\t')
```

```python
x = df['review']
y = df['sentiment']
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2)
```

```python
tfidf = TfidfVectorizer(analyzer=clean_text,max_features=5000)
```

```python

x_train = tfidf.fit_transform(x_train)
```

```python
x_test = tfidf.transform(x_test)
```

#### train a random forest model using our data

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rf_classifier.fit(x_train, y_train)
```

```python
y_pred = rf_classifier.predict(x_test)
```

#### get the accuracy of the model

```python
print("Accuracy: ", accuracy_score(y_test, y_pred))
```

#### create a confusion matrix of the model

```python
confusion = confusion_matrix(y_test, y_pred)
```

#### get the precision, recall, f1 of the model

```python
precision_score(y_test, y_pred)
```

```python
 recall_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred)
```

#### Create a Roc curve

```python
y_pred_proba = rf_classifier.predict_proba(x_test)[::,1]
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
```

```python
plt.plot(fpr,tpr)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
